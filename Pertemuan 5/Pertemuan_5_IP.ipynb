```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image


# 1. Load Citra

img = Image.open('/content/dog.jpg').convert('L') # Convert ke Grayscale
img_array = np.array(img)

# 2. Definisi Kernel Rataan 5x5 (Mean Filter)
# Sesuai slide: nilai tiap elemen adalah 1/25
kernel_size = 5
kernel = np.ones((kernel_size, kernel_size)) / (kernel_size**2)

def manual_convolution(image, kernel):
    img_h, img_w = image.shape
    ker_h, ker_w = kernel.shape

    # Menentukan ukuran output (tanpa padding)
    output_h = img_h - ker_h + 1
    output_w = img_w - ker_w + 1
    output = np.zeros((output_h, output_w))

    # Proses Konvolusi Manual
    for i in range(output_h):
        for j in range(output_w):
            # Ambil area sub-matriks sesuai ukuran kernel
            region = image[i:i+ker_h, j:j+ker_w]
            # Operasi perkalian elemen dan penjumlahan
            output[i, j] = np.sum(region * kernel)

    return output

# 3. Eksekusi Fungsi
output_img = manual_convolution(img_array, kernel)

# 4. Visualisasi Hasil
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title("Citra Asli (Grayscale)")
plt.imshow(img_array, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Hasil Konvolusi (Kernel Rataan 5x5)")
plt.imshow(output_img, cmap='gray')
plt.axis('off')

plt.show()
```
